```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
import sklearn.linear_model


#Read file
psearch_df = pd.read_csv('Paid Search - PPC Prediction Analysis - May 2016.csv')
```

```python
psearch_df.head()
```

```python
psearch_df = psearch_df.join(pd.get_dummies(psearch_df['Type'],prefix="Type"))
```

```python
psearch_df = psearch_df.join(pd.get_dummies(psearch_df['Device'],prefix="Device"))
```

```python
psearch_df = psearch_df.join(pd.get_dummies(psearch_df['Geo'],prefix="Geo"))
```

```python
psearch_df = psearch_df.join(pd.get_dummies(psearch_df['Vendor'],prefix="Vendor"))
```

```python
psearch_df.describe()
```

```python
X = psearch_df.drop(['Clicks','Type','Device','Geo','Vendor','Channel'], axis = 1)
y = psearch_df['Clicks']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

rfr = RandomForestRegressor()
rfr.fit(X,y)
rfr_score = cross_val_score(rfr,X, y)
print rfr_score.mean()
```

```python
feature_importances_array = rfr.feature_importances_

# (one line) Put the features themselves in a list
features_array = X.columns.values 

# Turn these into a Data Frame
feature_importances_df = pd.DataFrame({'importances' : feature_importances_array,
                                       'features' : features_array})

feature_importances_df.sort(columns='importances', ascending=False)
```

```python
psearch_df_result = pd.read_csv('Paid Search - PPC Prediction Analysis - May 2016.csv')
```

```python
psearch_df_result['Clicks_predict'] = rfr.predict(X)
```

```python
pd.DataFrame(psearch_df_result.to_csv('Paid Search Predicted_May 2016.csv', index=False, header=False))
```
